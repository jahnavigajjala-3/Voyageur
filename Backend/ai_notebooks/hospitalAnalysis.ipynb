```python
import pandas as pd
import numpy as np
import os
from math import radians, sin, cos, sqrt, atan2
```

```python
try:
    df = pd.read_csv("../data/hospitals.csv")
    df.columns = df.columns.str.strip()
    print(f"Loaded {len(df)} hospitals")
except FileNotFoundError:
    print("Error: hospitals.csv not found in ../data/")
df = df.dropna(subset=['Latitude', 'Longitude', 'Rating'])
```

```python
df = df.sort_values('Rating', ascending=False)
df = df.drop_duplicates(subset=['Latitude', 'Longitude'], keep='first')
```

```python
df['City']     = df['City'].str.title().str.strip()
df['State']    = df['State'].str.title().str.strip()
df['District'] = df['District'].str.title().str.strip()

print(f"After cleaning: {len(df)} hospitals")
```

```python
district_summary = df.groupby(['District', 'State']).agg(
    total_hospitals  = ('id', 'count'),
    avg_rating       = ('Rating', 'mean'),
    best_hospital_city = ('City', 'first'),
    best_rating      = ('Rating', 'max')
).reset_index()

district_summary['avg_rating'] = district_summary['avg_rating'].round(2)
```

```python
processed_df = df[[
    'id', 'City', 'State', 'District',
    'Latitude', 'Longitude',
    'Rating', 'Number of Reviews'
]].copy()
```

```python
hospital_output = "../data/processed_hospitals.csv"
district_output = "../data/hospital_district_summary.csv"

if not os.path.exists(hospital_output):
    processed_df.to_csv(hospital_output, index=False)
    print(f"Saved: {hospital_output}")
else:
    print("Hospital file already exists, skipping.")

if not os.path.exists(district_output):
    district_summary.to_csv(district_output, index=False)
    print(f"Saved: {district_output}")
else:
    print("District summary already exists, skipping.")
```

```python
print(f"\nTotal hospitals: {len(processed_df)}")
print(f"Total districts covered: {len(district_summary)}")
print(f"\nTop 5 districts by avg rating:")
print(district_summary.sort_values('avg_rating', ascending=False).head())
```
