```python
# AI Prototype Notebook

# Purpose:
# - Test Gemini API
# - Experiment with prompts
# - Debug responses

# NOTE:
# Production logic is in ai_service.py

from google import genai
import os
from dotenv import load_dotenv

load_dotenv()

client = genai.Client(api_key=os.getenv("GEMINI_API_KEY"))

# Try multiple models because availability/quota is model-specific

models = [
    "gemini-2.5-flash",
    "gemini-2.0-flash",
    "gemini-1.5-flash",
    "gemini-flash-latest"
]

response = None
last_error = None

for m in models:
    try:
        response = client.models.generate_content(
            model=m,
            contents="I just started my trip to Mumbai, any tips?"
        )
        print("Using model:", m)
        break
    except Exception as e:
        last_error = e
        print("Failed model:", m, e)

if response:
    print(response.text)
else:
    print("All models failed:", last_error)
```
